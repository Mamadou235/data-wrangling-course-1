```python
#Chargement des librairy
import numpy as np
import pandas as pd
#Chargement du ficher de travail
source = "HDR 2007-2008 Table 10.xlsx"
df = pd.read_excel(source)
#Afficher les 12 première lignes
df.head(12)
```

```python
# Repositionner la ligne d'en-tête en supprimant les espaces avant
df = pd.read_excel(source, header=3)
df
```

```python
# Délimiter le fichier
# Supprimez d'abord toutes les colonnes (axe=1) sans valeurs
df = df.dropna(how="all", axis=1)
# Supprimer toutes les lignes (axe=0) sans valeurs
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(6)
```

```python
#lister des informations qui sont déjà là et utilisons cela pour guider les nouveaux noms que nous devons inclure.
columns = []
for c in df.columns:
    # Pour chaque colonne, obtenir les valeurs des 3 premières lignes
    # puis conservez toutes les chaînes (c'est-à-dire excluez-les si elles sont nulles) et ajoutez-les en colonne
    # A partir de High humain dev fusionner les colonnes ou les champs sont non vides
    columns.append([str(v) for v in df[c][:3] if pd.notnull(v)])
columns
```

```python
#Affecter des nouvelles entêtes aux colomns plus précises et pour les columns sans entêtes mettre reference x
columns = [
    "HDI rank",
    "Country",
    "Life expectancy at birth(years) 1970-1975d",
    "Reference 1",
    "Life expectancy at birth(years) 2000-2005d",
    "Reference 2",
    "Infant mortality rate(per 1,000 live births) 1970",
    "Reference 3",
    "Infant mortality rate(per 1,000 live births) 2005",
    "Reference 4",
    "Under-five mortality rate(per 1,000 live births) 1970",
    "Reference 5",
    "Under-five mortality rate(per 1,000 live births) 2005",
    "Reference 6",
    "Probability at birth of not surviving to age 65a (% of cohort) Female 2000-2005d",
    "Reference 7",
    "Probability at birth of not surviving to age 65a (% of cohort) Male 2000-2005d",   
    "Reference 8",
    "Maternal mortality ratio\n (per 100,000 live births) Reportedb 1990-2005e",   
    "Reference 9",
    "Maternal mortality ratio\n (per 100,000 live births) Adjustedc 2005"
]
df
```

```python
# Réinitialiser les colonnes du jeu de données
df.columns = columns
df.head()
```

```python
# Supprimer les lignes qui contenaient les informations dont nous n'avons plus besoin
# et réinitialiser l'index
# np.arange(3) crée un tableau = [0, 1, 2, 3]
df = df.drop(np.arange(3), axis=0).reset_index(drop=True)
df
```

```python
#Afficher les 25 dernières lignes afin d'afficher les notes de bas de pages
df.tail(25)
```

```python
#Sauvegarder les notes,sources et métadonnées dans des dataframes
df_meta = df.iloc[180:197,]
df_notes = df.iloc[197:199,]
df_src = df.iloc[199:200,]


df_meta = df_meta.dropna(how = 'any', axis = 1).reset_index(drop=True)
df_notes = df_notes.dropna(how = 'any', axis = 1).reset_index(drop=True)
df_src = df_src.dropna(how = 'any', axis = 1).reset_index(drop=True)
```

```python
# df sans les notes, sources et métadonnées
df = df.drop(np.arange(179,len(df)), axis=0).reset_index(drop=True)
df
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]

# Se débarrasser des rangées avec les catégories HDI
df = df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)

# Référence pour np.where https://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html
# Pensez que cela ressemble à une déclaration du type "if" : where(test est vrai, vrai-quelque chose, faux-quelque chose)
# ceci peut aussi être imbriqué : where(test is true, true-thing, where(new-test is true, true-thing, false-thing))
# Classification des Pays selon le niveau de developpement HDI rank 
df.loc[:, "HDI category"] = np.where(df["HDI rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI rank"] > 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI rank", inplace=True)
df
```

```python
#Enregistrons nos données

with pd.ExcelWriter('HDR 2007-2008 Table 10.xlsx') as writer:
    df.to_excel(writer, sheet_name = 'Table_10', index = False)
    df_notes.to_excel(writer, sheet_name = 'NOTES', index = False)
    df_src.to_excel(writer, sheet_name = 'SOURCES', index = False)
    df_meta.to_excel(writer, sheet_name = 'METADONNEES', index = False)
```
